#### Анализ грузопотоков 

На железной дороге, как и в других логистических экосистемах, немалое внимание уделяется анализу грузопотоков.  
Грузовой поток - это количество груза, перевозимого в одном направлении за определенный период времени, где направлением считается каждая уникальная пара пункт отправления - пункт назначения.  
Каждый грузовой поток характеризуется несколькими параметрами:  
* Кодом отправителя
* Кодом пункта отправления
* Кодом пункта назначения
* Минимально возможным объемом перевозимого груза
* Максимально возможным объемом перевозимого груза

Внутри ПГК задача по анализу грузопотоков лежит на плечах Олеси и Влада.
Схема их работы такова: Олеся у каждого отправителя узнаёт объем груза $n$, который будет передан к перевозке и сообщает Владу.
Задачей Влада является распределить значение, полученное от Олеси по направлениям, соблюдая следующие условия:  
> Если значение $n$, указанное Олесей ниже чем минимально возможный объем $min$, то $n$ распределяется по направлениям пропорционально минимально возможному объему $min$  
> Если значение $n$ выше чем минимально возможный объем $min$, то $n$ сначала распределяется на направления по которым есть $min$, а остаток свыше минимально возможного объема $n-min$ распределяется по направлениям пропорционально максимально возможному объему $max-min$
